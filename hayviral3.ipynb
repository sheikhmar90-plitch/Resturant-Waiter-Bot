# PYTHON + AI CONTEXT PROMPT ENGINEERING(RESTURANT WAITER BOT)

```python

from difflib import get_close_matches

# ================= RESTAURANT NAME =================
RESTAURANT_NAME = "Royal Taste Restaurant"

# ================= SYSTEM PROMPT =================
SYSTEM_PROMPT = """
You are a strict conversational automation waiter proxy agent.
You must only identify items, reference indexes, and query validations native to the active dictionary constraints.
Strict limit: Do not hallucinate external entries.
"""

print(SYSTEM_PROMPT)

# ================= MENU =================
menu = {
    "Chicken Pizza": 1200,
    "Beef Burger": 650,
    "Club Sandwich": 450,
    "Fries": 250,
    "Pasta" : 350,
    "fried chicken" : 450,
    "Water" : 70,
    "Coke" : 50,
    "Icecream" : 120
}

special_items = {
    "Chicken Pizza Combo": 1500,
    "Family Burger Deal": 2000,
    "Family BBQ Deal" : 2400
}

# ================= ORDER STORAGE =================
order = {}

# ================= SMART MATCH FUNCTION =================
def find_item(user_input, items):
    user_input = user_input.lower().strip()

    matches = get_close_matches(
        user_input,
        [i.lower() for i in items],
        n=1,
        cutoff=0.6
    )

    if matches:
        for item in items:
            if item.lower() == matches[0]:
                return item
    return None


# ================= SAFE INPUT =================
def safe_input(prompt):
    try:
        return input(prompt).strip()
    except:
        return ""


# ================= START =================
print(f"\n🍽️ Welcome to {RESTAURANT_NAME}!")
print("How can I help you today? ")

while True:

    print("\n----- MENU OPTIONS -----")
    print("1. Show Menu")
    print("2. Show Special Items")
    print("3. Order Item")
    print("4. View Bill")
    print("5. Exit")

    choice = safe_input("Enter choice: ").lower()

    # 1 MENU
    if choice.startswith("1"):
        print(f"\n📋 {RESTAURANT_NAME} MENU:")
        for item, price in menu.items():
            print(f"{item} - Rs.{price}")

    # 2 SPECIAL
    elif choice.startswith("2"):
        print("\n TODAY'S SPECIAL:")
        for item, price in special_items.items():
            print(f"{item} - Rs.{price}")

    # 3 ORDER
    elif choice.startswith("3"):

        item_input = safe_input("Enter item name: ")
        item = find_item(item_input, list(menu.keys()) + list(special_items.keys()))

        if item:

            qty_input = safe_input("Enter quantity: ")

            if qty_input.isdigit():
                qty = int(qty_input)

                price = menu.get(item, special_items.get(item))

                if item in order:
                    order[item]["qty"] += qty
                else:
                    order[item] = {"price": price, "qty": qty}

                print(f"✔ {item} x{qty} added to order")

            else:
                print(" Invalid quantity")

        else:
            print(" Item not found (try again)")

    # 4 BILL
    elif choice.startswith("4"):

        if not order:
            print(" No items ordered yet")

        else:
            print(f"\n {RESTAURANT_NAME} BILL RECEIPT")
            print("----------------------------")

            total = 0

            for item, data in order.items():
                item_total = data["price"] * data["qty"]
                total += item_total
                print(f"{item} x{data['qty']} = Rs.{item_total}")

            discount = 0
            if total > 2000:
                discount = total * 0.1

            print("\n----------------------------")
            print(f"Subtotal: Rs.{total}")
            print(f"Discount: Rs.{discount}")
            print(f"Total Payable: Rs.{total - discount}")

    # 5 EXIT
    elif choice.startswith("5") or choice == "exit":
        print(f" Thank you for visiting {RESTAURANT_NAME}!")
        break

    else:
        print(" Invalid input, please choose 1-5")
```

# summary

**This Python-based waiter chatbot is a rule-based restaurant management system that simulates real-world food ordering. It allows users to view a menu, select items, place orders with quantity, and generate a final bill. The system uses a dictionary-based structure to store menu items and prices, and it applies fuzzy string matching (using difflib.get_close_matches) to handle typos and improve user experience. Input normalization techniques like .strip() and .lower() ensure flexible and clean input handling. When a user places an order, the system stores each item along with its price and quantity in a dictionary. The billing system then calculates the total cost by multiplying price and quantity, applies a discount if the total exceeds a defined threshold, and generates a structured receipt with restaurant branding. A system prompt is included to enforce strict rules, ensuring the chatbot only uses predefined menu items and does not generate invalid or external data. Overall, this project demonstrates input handling, error management, fuzzy matching, and dynamic billing in a simple but realistic restaurant simulation system.**
